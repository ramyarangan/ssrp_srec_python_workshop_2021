```python
# a standalone piece of code
import pandas as pd
import matplotlib.pyplot as plt

nytimes_covid_live_url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-counties.csv'

df = pd.read_csv(nytimes_covid_live_url)
groups_bystate = df.groupby('state')
df_states = groups_bystate.sum().sort_values('cases', ascending=True)
df_states_20 = pd.DataFrame(df_states.head(20))
df_states_20['fr'] = df_states_20['deaths'] / df_states_20['cases'] * 100

df_states_20.plot.scatter(x='cases',y='fr', color='red')
plt.ylabel('FR')
plt.title('A cool plot')

df_states_20.sort_values('fr', ascending=False).plot.bar(y='fr', color='blue')
plt.title('A cooler plot')
```
