A very simple introduction to matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[8,6,1,3,8,0,1]

plt.plot(x,y)
```

We can fine tuning the plot

```python
plt.plot(x,y, color='red', linewidth=4, marker="x", markersize=20, linestyle=":")
plt.title('A coffee study')
plt.xlabel('Day of the week')
plt.ylabel('Numer of coffe cups drunk')
```

Plot mutiple datasets

```python
y2=[1,1,1,1,1,1,0]
plt.plot(x,y, color='red', linewidth=4, marker="x", markersize=20, linestyle=":", label='researcher1')
plt.plot(x,y2, color='blue', linewidth=4, marker="x", markersize=20, linestyle=":", label='researcher2')
plt.title('A coffee study')
plt.xlabel('Day of the week')
plt.ylabel('Numer of coffe cups drunk')

plt.legend()
```

Add error bars

```python
yerr=[1,2,1,0,0,1,0]
plt.errorbar(x,y, yerr=yerr, color='red', linewidth=4, marker="x", markersize=20, linestyle=":", label='researcher1')
#plt.errorbar(x,y, yerr=yerr, color='black', linewidth=1, marker="x", markersize=20)
plt.title('A coffee study')
plt.xlabel('Day of the week')
plt.ylabel('Numer of coffe cups drunk')
plt.legend()
```

Scatter plot

```python
plt.scatter(x,y, label='researcher1')
plt.scatter(x,y2, label='researcher2')
plt.title('A coffee study')
plt.xlabel('Day of the week')
plt.ylabel('Numer of coffe cups drunk')
plt.legend()
```


Bar plot

```python
import numpy as np
```

```python
xpos=np.arange(len(y))
w=0.3
plt.bar(xpos-w/2,y, width=0.3)
plt.bar(xpos+w/2,y2, width=0.3)
```

Multiple plots in a grid

```python
fig, axs=plt.subplots(1,2)
axs[0].bar(xpos, y, width=0.6)
axs[1].bar(xpos, y2, width=0.6)

for ax in axs:
  ax.set_xlabel('Day of the week')
  ax.set_ylabel('Number of coffee cups')

plt.tight_layout()
```
